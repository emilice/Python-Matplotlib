```python
import numpy as np
t=np.arange(0,5,0.2) #0'dan 5'e 0.2 artış sergilesin, bir dizi oluşturur
plt.plot(t,t,'g--',t,t**2,'b*') #b-blue, y-yellow, g-green...
```

```python
gruplar=["A","B","C"]
veriler=[15,45,30]

plt.figure(figsize=(9,3))
plt.subplot(131)

plt.bar(gruplar,veriler)
plt.subplot(132)
plt.scatter(gruplar,veriler)

plt.subplot(133)
plt.plot(gruplar, veriler)

plt.suptitle("Matplotlib Kütüphane Örnekleri")
```
